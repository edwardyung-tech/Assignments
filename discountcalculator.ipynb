```python
# asking user to enter total number of items
quantity_of_items = int (input ("Enter total number of items purchased: "))
#calculating total amount without any discount 
amount_of_each_item = 10
total_amount = quantity_of_items * amount_of_each_item
print (f"Your total amount is {total_amount} $")
#asking the user if he/she is a premium_member
premium_membership = input("Are you a premium member ? (yes/no) :")
#calculating discount based on amount of items bought 
if quantity_of_items > 10  :
    print (f"You are eligible for a 10% discount of {total_amount*0.1}$")
    discount_price = total_amount - (total_amount*0.1)
    print(f"Your total amount is now: {discount_price }$")
#calculating discount based on premium membership status 
    if premium_membership == "yes":
        print ("Congratulations you are also eligible for a discount of 5$ for being a premium member ")
        premium = discount_price - 5
        print (f"Total amount including 10% discount and premium member ship is now :{premium}$")
    else :
        print (f"Your total amount is now: {discount_price }$")
else :
    print ("No discount can be allowed ")
    print (f"Total amount is {total_amount}$")
        

```
