```python
# asking the user to input the score to be graded 
score = int(input ("Enter your score: "))
# determining if the score is a valid score between 1 and 100 and determine grades A1,B2,B3,C4,C5,C6,D7,E8 or F9 using conditional statements 
if (score < 0 or score > 100):
  print ("Enter a valid score between 0 and 100")
else:
  if (score >= 75 ):
    print("Your grade is A1")
  elif (score >= 70 and score <75):
    print ("Your grade is B2")
  elif (score >= 65 and score < 70 ):
    print ("Your grade is B3")
  elif (score >= 60 and score < 65):
    print ("Your grade is C4")
  elif (score >= 55 and score < 60):
    print ("Your grade is C5")
  elif (score >= 50 and score < 55):
    print ("Your grade is C6")
  elif (score >= 45 and score < 50):
    print ("Your grade is D7")
  elif (score >= 40 and score < 45):
    print ("Your grade is E8")
  elif (score >= 0 and score < 40):
    print ("Your grade is F9")
```
